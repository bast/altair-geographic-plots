```python
import altair as alt
import pandas as pd
from vega_datasets import data

cities = pd.read_csv("cities.csv")

cities.head()
```

```python
countries = alt.topo_feature(data.world_110m.url, "countries")
world = alt.Chart(countries).mark_geoshape(fill="lightgray", stroke="gray")

points = (
    alt.Chart(cities)
    .mark_point(size=50, color="blue")
    .encode(
        latitude="lat",
        longitude="lon",
    )
)

chart = (
    (world + points)
    .properties(width=500, height=500)
    .project(
        type="equalEarth",
        scale=800,
        translate=[100, 1000],
    )
)

chart
```
